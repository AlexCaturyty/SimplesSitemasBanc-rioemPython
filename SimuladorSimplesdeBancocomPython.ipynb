```python
menu = """

[d] Depositar
[s] Sacar
[e] Extrato
[q] Sair

=> """

saldo_da_conta = 0
limite_da_conta = 500
extrato_da_conta = ""
numero_saques_da_conta = 0
LIMITE_SAQUES = 3

while True:

    opcao = input(menu)

    if opcao == "d":
        valor = float(input("Informe o valor do depósito: "))

        if valor > 0:
            saldo_da_conta += valor
            extrato_da_conta += f"Depósito: R$ {valor:.2f}\n"

        else:
            print("O valor informado é inválido.")

    elif opcao == "s":
        valor = float(input("Informe o valor do saque: "))

        excedeu_saldo_da_conta = valor > saldo_da_conta

        excedeu_limite_da_conta = valor > limite_da_conta

        excedeu_saques_da_conta = numero_saques_da_conta >= LIMITE_SAQUES

        if excedeu_saldo_da_conta:
            print("Você não tem saldo suficiente.")

        elif excedeu_limite_da_conta:
            print("O valor do saque excede o limite.")

        elif excedeu_saques_da_conta:
            print("Número máximo de saques excedido.")

        elif valor > 0:
            saldo_da_conta -= valor
            extrato_da_conta += f"Saque: R$ {valor:.2f}\n"
            numero_saques_da_conta += 1

        else:
            print("O valor informado é inválido.")

    elif opcao == "e":
        print("\n================ EXTRATO ================")
        print("Nenhum depósito ou saque foi realizado." if not extrato_da_conta else extrato_da_conta)
        print(f"\nSaldo: R$ {saldo_da_conta:.2f}")
        print("==========================================")

    elif opcao == "q":
        break

    else:
        print("Operação inválida, por favor selecione novamente a operação desejada.")
```
